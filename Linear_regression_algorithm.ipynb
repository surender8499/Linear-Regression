# Build your own Linear Regression Algorithm 

```python
from statistics import mean
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xd = [1,2,3,4,5,6,7,8]
yd = [5,4,6,5,6,7,8,9]
plt.scatter(xd,yd)
plt.show()
```

```python
xd = np.array([1,2,3,4,5,6,7,8], dtype=np.float64)
yd = np.array([5,4,6,5,6,7,8,9], dtype=np.float64)
```

```python
def best_fit_slope(xd,yd):
    m= (((mean(xd)*mean(yd)) - mean(xd*yd)) / 
        ((mean(xd)**2) - mean(xd*xd)))
    return m
```

```python
m = best_fit_slope(xd,yd)
```

```python
print(m)
```

```python
xs = [1,2,3,4,5]
ys = [5,4,6,5,6]
xs = np.array([1,2,3,4,5], dtype=np.float64)
ys = np.array([5,4,6,5,6], dtype=np.float64)
def best_fit_slope1(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)**2) - mean(xs**2)))
    return m
```

```python
m = best_fit_slope1(xs,ys)
print(m)
```

```python
def best_fit_slope_and_intercept(xs,ys):
    m = (((mean(xs)*mean(ys)) - mean(xs*ys)) /
         ((mean(xs)**2) - mean(xs**2)))
    b = mean(ys) - m*mean(xs)
    return m, b

```

```python
m, b = best_fit_slope_and_intercept(xs,ys)
```

```python
print(m,b)
```

```python
regression_line = [(m*x + b) for x in xs]
```

```python
def squared_error(ys_orig, ys_line):
    return sum((ys_line - ys_orig)**2)

def coefficient_of_determination(ys_orig,ys_line):
    y_mean_line = [mean(ys_orig) for y in ys_orig]
    squared_error_regr = squared_error(ys_orig, ys_line)
    squared_error_y_mean = squared_error(ys_orig, y_mean_line)
    return 1 - (squared_error_regr/squared_error_y_mean)
```

```python
r_squared = coefficient_of_determination(yd,regression_line)
print(r_squared)
```

```python
m, b = best_fit_slope_and_intercept(xs,ys)
regression_line = [(m*x + b) for x in xs]
r_squared = coefficient_of_determination(ys,regression_line)
print(r_squared)
```

```python
#Making a dataset with random values
import random
def create_dataset(hm,variance,step=2,correlation=False):
    val = 1
    ys = []
    for i in range(hm):
        y = val + random.randrange(-variance,variance)
        ys.append(y)
        if correlation and correlation == 'pos':
            val+=step
        elif correlation and correlation == 'neg':
            val-=step

    xs = [i for i in range(len(ys))]
    
    return np.array(xs, dtype=np.float64),np.array(ys,dtype=np.float64)
```

```python
new_xs, new_ys = create_dataset(40,40,2,correlation='pos')
```

```python
m, b = best_fit_slope_and_intercept(new_xs,new_ys)
regression_line = [(m*x + b) for x in new_xs]
plt.scatter(new_xs,new_ys)
plt.plot(new_xs, regression_line)
plt.show()
```

```python
r_squared = coefficient_of_determination(new_ys,regression_line)
print(r_squared)
```
